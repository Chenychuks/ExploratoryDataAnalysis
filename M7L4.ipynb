```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('house_pricing_dataset.csv')
df
```

```python
df.price.describe()
```

```python
#The distribution of house prices using the 1st and 2nd perscentile
plt.hist(df['price'], range=(322875.0, 460943.5), bins=30, color = 'green')
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
plt.title('The distribution of house prices', fontdict = font1)
plt.xlabel('House price', fontdict = font2)
plt.ylabel('Distribution', fontdict = font2)
plt.show()
```

```python
#The distribution of house prices from the minimum to the maximum house price 
plt.hist(df['price'], range=(0.000000e+00,6.549625e+05), bins=20, color = 'brown')
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
plt.title('The distribution of house prices', fontdict = font1)
plt.xlabel('House price', fontdict = font2)
plt.ylabel('Distribution', fontdict = font2)
plt.show()
```

```python
# The distribution of the years the houses were last renovated
sns.kdeplot(df['yr_renovated'], shade=True)
plt.show()
```

```python
#The relationship between prices and number of bedrooms
plt.scatter(df.bedrooms, df.price, color = 'maroon', marker = '*')

plt.title('relationship between prices and number of bedrooms'.upper(), fontsize = 15)
plt.xlabel('BEDROOMS', fontdict = font2)
plt.ylabel('PRICE', fontdict = font2)
plt.show()
```

```python
# The relationship between number of bedrooms and year the house was built
plt.scatter(df.yr_built, df.bedrooms, color = 'maroon', marker ='*')
plt.title('relationship between number of bedrooms and year the house was built'.upper(), fontsize = 10)
plt.xlabel('YEAR BUILT', fontdict = font2)
plt.ylabel('BEDROOM', fontdict = font2)
plt.show()
```

```python
#The distribution of average house prices across the top 10 cities
sales_mean = df.groupby(by='city', as_index=False)['price'].mean()
Reset = sales_mean.reset_index()

Top_city = Reset.sort_values('city', ascending = False)
Top_city_df = Top_city.head(10)
Top_city_df

plt.figure(figsize=(15,9))
plt.bar(Top_city_df['city'], Top_city_df['price'])
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
plt.title('Distribution of average house prices across the top 10 cities'.upper(), fontdict = font1)
plt.ylabel('PRICE', fontdict = font2)
plt.xlabel('CITY', fontdict = font2)
plt.grid()
plt.show()
```

```python
# The distribution of maximum bedroom size across the top 10 cities

Size = df.groupby(by='city', as_index=False)['bedrooms'].sum()
Reset = Size.reset_index()

Bedroom_size = Reset.sort_values('city', ascending = False)
Top_size_df = Bedroom_size.head(10)
Top_size_df

plt.figure(figsize=(15,9))
plt.bar(Top_size_df['city'], Top_size_df['bedrooms'])
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
plt.title('Distribution of maximum bedroom size across the top 10 cities'.upper(), fontdict = font1)
plt.ylabel('BEDROOMS', fontdict = font2)
plt.xlabel('CITY', fontdict = font2)
plt.grid()
plt.show()
```

```python
# The distribution of average basement square feets across top 10 cities
Basement_mean = df.groupby(by='city', as_index=False)['sqft_basement'].mean()
Reset = Basement_mean.reset_index()

Basement = Reset.sort_values('city', ascending = False)
Basement_df = Basement.head(10)
Basement_df

plt.figure(figsize=(15,9))
plt.bar(Basement_df['city'], Basement_df['sqft_basement'])
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
plt.title('Distribution of average basement square feets across top 10 cities'.upper(), fontdict = font1)
plt.ylabel('BASEMENT BY SQUARE FEETS', fontdict = font2)
plt.xlabel('CITY', fontdict = font2)
plt.grid()
plt.show()
```

```python
# The trend of house prices from the earliest date to the latest date in the dataset
df1= df[['date', 'price']]
# convert the date column to a datetime object
df1['date'] = df['date'].apply(pd.to_datetime)

# group the weekly sales by date
Price_sum = df1.groupby(by='date', as_index=False)['price'].sum()
plt.plot(Price_sum['date'], Price_sum['price'])
font1 = {'family':'serif','color':'black','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
plt.title('trend of house prices'.upper(), fontdict = font1)
plt.xlabel('Year'.upper(), fontdict = font2)
plt.ylabel('PRICE', fontdict = font2)
plt.show()
```
